```python
#line plot
# y dependent on x
# x independent
# tempreature get effected by the time

hours = ["6 AM", "9 AM", "12 PM", "3 PM", "6 PM", "7 PM" ]
tempreature = ['20C', '25C', '30C', '40C', '35C', '20C',]

plt.plot(hours, tempreature, color='red', marker='o')
plt.title("hours vs tempreature")
plt.xlabel("Hours")
plt.ylabel("Tempreature")
plt.grid(True)
plt.show()
```

```python
#Bar chart
products=['Laptops', 'Phones', 'Tablets']
sales = [200, 300, 100]
plt.bar(products, sales, color = 'purple',  edgecolor='black', linewidth=1)
plt.title("sales by product category", fontsize=14, fontweight='bold')
plt.xlabel("Products", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#histogram
scores = [55, 67, 72, 73, 68, 90, 88, 75, 82, 60, 59, 95, 77, 62, 83]

plt.figure(figsize=(6, 4))
plt.hist(scores, bins=5, color='pink', edgecolor='black')

plt.title("Student Exam Score Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Score Range", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
data = np.random.randn(1000)

# List of possible bin sizes
bin_sizes = [10, 20, 30, 50, 100]

# Plot histograms with different bin sizes
plt.figure(figsize=(12, 8))

for i, bins in enumerate(bin_sizes, 1):
    plt.subplot(2, 3, i)
    plt.hist(data, color='purple', bins=bins, edgecolor='black', alpha=0.7)
    plt.title(f"Bins = {bins}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot style and color palette
sns.set(style="whitegrid", palette="muted")

# Build dataset with different distributions
data = pd.DataFrame({
    'type': ['variable 1'] * 1000 + ['variable 2'] * 1000,
    'value': np.concatenate([np.random.randn(1000), np.random.normal(loc=4, size=1000)])
})

# Create the histogram with small multiples (facets) for different 'type'
plt.figure(figsize=(12, 6))

g = sns.FacetGrid(data, col="type", height=5, aspect=1.5)
g.map(plt.hist, 'value', bins=30, color='skyblue', edgecolor='black', alpha=0.6)

# Set axis labels and titles
g.set_axis_labels("Value", "Frequency")
g.set_titles("{col_name}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Generate sample data
data = pd.DataFrame({
    'type': ['variable 1'] * 1000 + ['variable 2'] * 1000,
    'value': np.concatenate([np.random.randn(1000), np.random.normal(loc=4, size=1000)])
})

# Initialize the matplotlib figure
plt.figure(figsize=(8, 6))

# Plot histogram with transparency (alpha)
for label, color in zip(['variable 1', 'variable 2'], ['#8dd3c7', '#8073ac']):
    subset = data[data['type'] == label]
    plt.hist(subset['value'], bins=30, alpha=0.7, label=label, color=color, edgecolor='white')

# Customize plot
plt.xlabel('')
plt.ylabel('')
plt.legend(title='', loc='upper right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

digit = np.random.rand(8, 8)

fig, axs = plt.subplots(2, 3, figsize=(12, 6))

axs[0, 0].imshow(digit, cmap='viridis', interpolation='nearest')
axs[0, 0].set_title('viridis')

axs[0, 1].imshow(digit, cmap='plasma', interpolation='nearest')
axs[0, 1].set_title('plasma')

axs[0, 2].imshow(digit, cmap='inferno', interpolation='nearest')
axs[0, 2].set_title('inferno')

axs[1, 0].imshow(digit, cmap='magma', interpolation='nearest')
axs[1, 0].set_title('magma')

axs[1, 1].imshow(digit, cmap='cividis', interpolation='nearest')
axs[1, 1].set_title('cividis')

axs[1, 2].axis('off')


plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Region A Sales': [200, 220, 250, 270, 300, 310, 400, 420, 450, 470, 500, 550],
    'Region B Sales': [150, 180, 190, 200, 230, 250, 300, 320, 340, 360, 400, 430],
    'Region C Sales': [100, 120, 140, 150, 160, 180, 190, 210, 220, 230, 250, 280],
    'Advertising Budget': [50, 60, 55, 65, 70, 75, 85, 90, 95, 100, 110, 120]
}

df = pd.DataFrame(data)
print(df)

#line plot
plt.figure(figsize = (5,5))
listofRegions = ['Region A Sales', 'Region B Sales', 'Region C Sales']
for region in listofRegions:
  plt.plot(df['Month'], df[region], label = region)

plt.title('Monthly sales trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()

#bar chart
total_sales = df[['Region A Sales', 'Region B Sales', 'Region C Sales']].sum()
total_sales.plot(kind='bar', figsize=(5,5), color=['blue', 'orange', 'green'])
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()

#histogram
plt.figure(figsize=(5,5))
plt.hist(df['Region A Sales'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Region A Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

#scatter plot
plt.figure(figsize=(5,5))
plt.scatter(df['Advertising Budget'], df['Region A Sales'], color='purple', edgecolor='black')
plt.title('Advertising Budget vs. Region A Sales')
plt.xlabel('Advertising Budget')
plt.ylabel('Region A Sales')
plt.show()

#pie chart
total_sales.plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green'])
plt.title('Proportion of Total Sales by Region')
plt.ylabel('')
plt.show()

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
print(df.head())
print(df.info())
```

```python
df_numeric = df.select_dtypes(include=["number"])
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap - Iris Dataset")
plt.show()
```
