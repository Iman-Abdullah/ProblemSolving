```python
from datetime import datetime, date, time, timedelta

data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%Y')
print(data['Date'].dtype)
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Weekday'] = data['Date'].dt.day_name()

print(data[['Date', 'Year', 'Month', 'Day', 'Weekday']])
```

```python
df_1985 = data[data['Year'] == 1985]

below_10 = df_1985[df_1985['Daily minimum temperatures'] < 10]
print(f"Number of days with temperature below 10°C: {len(below_10)}")
```

```python
avg = data.groupby('Year')['Daily minimum temperatures'].mean()

maxyear = avg.idxmax()
minyear = avg.idxmin()

print(f"Highest average temperature : {maxyear} , {avg[maxyear]}°C")
print(f"Lowest average temperature : {minyear} , {avg[minyear]}°C")
```

# Summary

The (Daily Minimum Temperatures) dataset was used to analyze time-related features like year, month, and day of the week. In 1985, it was found that some days had temperatures below 10°C, indicating a colder period. The exercise showed the power of pandas and Python's datetime tools in analyzing time series data and finding patterns.
