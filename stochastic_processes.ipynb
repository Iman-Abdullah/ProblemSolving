Scenario
You are modeling the arrival of customers at a bank. The time between customer arrivals follows an exponential distribution with an average arrival rate of 3 customers per hour (i.e., Î» = 3).

You want to:

Simulate the arrival times of the first 10 customers.

Plot the arrival times on a timeline.

Calculate the average time between arrivals from the simulated data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_rate = 3  # customers per hour
num_customers = 10

# 1. Simulate inter-arrival times (exponential distribution)
inter_arrival_times = np.random.exponential(scale=1/lambda_rate, size=num_customers)

# 2. Compute actual arrival times (cumulative sum)
arrival_times = np.cumsum(inter_arrival_times)

# 3. Print results
print("Inter-arrival times (in hours):", inter_arrival_times)
print("Arrival times (in hours):", arrival_times)

# 4. Plotting arrival times
plt.figure(figsize=(10, 4))
plt.step(range(1, num_customers+1), arrival_times, where='post', marker='o')
plt.title("Customer Arrival Times (Poisson Process)")
plt.xlabel("Customer Number")
plt.ylabel("Arrival Time (hours)")
plt.grid(True)
plt.show()

# 5. Compute and print average inter-arrival time
average_inter_arrival = np.mean(inter_arrival_times)
print(f"Average inter-arrival time: {average_inter_arrival:.2f} hours")
```
