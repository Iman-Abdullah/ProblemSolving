```python
from scipy.stats import norm
import math
import numpy as np
import matplotlib.pyplot as plt

# Confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Get z-value for two-tailed test
z = norm.ppf(1 - alpha / 2)

# Given data
mean = 3.2
sigma = 1
n = 30

# Standard error
SE = sigma / math.sqrt(n)

# Confidence interval
lower = mean - z * SE
upper = mean + z * SE

print(f"Z-value: {z:.4f}")
print(f"95% Confidence Interval: ({lower:.4f}, {upper:.4f})")

# Plot
x = np.linspace(mean - 4*SE, mean + 4*SE, 1000)
y = norm.pdf(x, mean, SE)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sample Mean Distribution', color='blue')
plt.fill_between(x, y, where=(x >= lower) & (x <= upper), color='orange', alpha=0.5, label='95% Confidence Interval')
plt.axvline(mean, color='blue', linestyle='--', label='Sample Mean = 3.2')
plt.title("Confidence Interval for Ice Cream Consumption (Known Standard Deviation)")
plt.xlabel("Ice Cream Scoops per Week")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

# Given values
p_hat = 60 / 200
n = 200
z = 1.645  # z-score for 90% confidence level

# Standard Error
se = math.sqrt(p_hat * (1 - p_hat) / n)

# Confidence Interval
lower = p_hat - z * se
upper = p_hat + z * se

print("90% Confidence Interval: ({:.3f}, {:.3f})".format(lower, upper))
```

```python
import math
from scipy.stats import norm

# Given data
mu = 80
x_bar = 78
sigma = 2.5
n = 40

# Calculate z-score
z = (x_bar - mu) / (sigma / math.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.3f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis. Machine is not working properly.")
else:
    print("Fail to reject the null hypothesis. Not enough evidence.")
```

```python
import math
from scipy.stats import t

# Given data
mu = 2
x_bar = 1.8
s = 0.15
n = 10
df = n - 1

# t-score
t_score = (x_bar - mu) / (s / math.sqrt(n))

# One-tailed p-value
p_value = t.cdf(t_score, df)

print(f"T-score: {t_score:.3f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.01:
    print("Reject the null hypothesis. The average lifespan is significantly less than 2 years.")
else:
    print("Fail to reject the null hypothesis.")
```
